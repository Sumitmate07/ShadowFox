```python
for i in df:
    if df[i].isnull().sum()>0:
        print(i)
        print(df[i].mean())
        print(df[i].median())
        
        print('\n')
```

```python




# Filling missing values based on skewness and data type
for i in df.columns:
    if df[i].isnull().sum() > 0:  # Check if column has missing values
        
        if df[i].dtype == 'object':  # Categorical data â†’ Use Mode
            df[i].fillna(df[i].mode()[0], inplace=True)
        elif df[i].dtype in ['int64', 'float64']:  # Numeric data
            skewness = df[i].skew()

            if skewness > 1 or skewness < -1:  # Highly skewed â†’ Use Median
                df[i].fillna(df[i].median(), inplace=True)
            else:  # Near normal distribution â†’ Use Mean
                df[i].fillna(df[i].mean(), inplace=True)


```

# 12.MODEL EVALUATION

```python
import sklearn
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
 
```

#Training Data

```python
def training_evaluation(actual,predicted):
    mse = mean_squared_error(actual,predicted)
    rmse= np.sqrt(mean_squared_error(actual,predicted))
    mae=mean_absolute_error(actual,predicted)
    r2=r2_score(actual,predicted)

    print(f'MSE :{mse}')
    print(f'RMSE :{rmse}')
    print(f'MAE :{mae}')
    print(f'R2_Score :{r2}')

    plt.scatter(actual,predicted,color='r')
    plt.plot([actual.min(),actual.max()],[actual.min(),actual.max()],color='k',lw=2,ls='--')
    plt.title('Best Fit Line-Training Data')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.show()
```

```python
training_evaluation(y_train,y_pred_train)
```

#Test Data

```python
def test_evaluation(actual,predicted):
    mse = mean_squared_error(actual,predicted)
    rmse= np.sqrt(mean_squared_error(actual,predicted))
    mae=mean_absolute_error(actual,predicted)
    r2=r2_score(actual,predicted)

    print(f'MSE :{mse}')
    print(f'RMSE :{rmse}')
    print(f'MAE :{mae}')
    print(f'R2_Score :{r2}')

    plt.scatter(actual,predicted,color='g')
    plt.plot([actual.min(),actual.max()],[actual.min(),actual.max()],color='k',lw=2,ls='--')
    plt.title('Best Fit Line-Test Data')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.show()
```

```python
test_evaluation(y_test,y_pred_test)
```
